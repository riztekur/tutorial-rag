```python
import os
from dotenv import load_dotenv
from llama_cloud_services import LlamaParse
```

```python
LLAMA_CLOUD_API_KEY = os.environ.get("LLAMA_CLOUD_API_KEY")
```

```python
parser = LlamaParse(
    api_key=LLAMA_CLOUD_API_KEY,
    verbose=True,
    # language='en' -> 'id','fr'
    # disable_ocr=True atau False
    # skip_diagonal_text = True atau false
    # target_pages="0,1,3" (index halaman yang akan diparsing)
    # disable_image_extraction = True atau False
)
```
