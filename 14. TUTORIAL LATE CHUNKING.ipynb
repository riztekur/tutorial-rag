```python
from chonkie import LateChunker, RecursiveChunker, RecursiveRules, RecursiveLevel
from pathlib import Path
```

```python
text = Path("documents/bandung.md").read_text(encoding="utf-8")
```

```python
rules = RecursiveRules(levels=[
    # Level 4: Paragraf (baris kosong / newline)
    RecursiveLevel(
        delimiters=["\n\n", "\r\n\r\n"],  # boleh pakai newline sebagai delimiter
        include_delim="prev",
    ),
    # Level 5: Kalimat (fallback)
    RecursiveLevel(
        delimiters=[". ", "! ", "? "],
        include_delim="prev",
    ),
    # Level 6: Whitespace kata (paling halus, jika masih terlalu panjang)
    RecursiveLevel(
        whitespace=True
    ),
])
```

```python
chunker = LateChunker(
    embedding_model = "all-MiniLM-L6-v2",
    chunk_size=256,
    rules=rules,
    min_characters_per_chunk=64
)
```

```python
chunks = chunker.chunk(text)
```
