```python
import os
from dotenv import load_dotenv

import voyageai
import PIL
```

```python
VOYAGE_API_KEY = os.environ.get("VOYAGE_API_KEY")
```

```python
vo = voyageai.Client(api_key=VOYAGE_API_KEY)
```

```python
# input

inputs = [
    [
        "Jalan Braga adalah salah satu jalan tertua di Bandung yang kini menjadi pusat kuliner, seni, dan suasana kota tua yang kental dengan kafe-kafe vintage.",
        PIL.Image.open('documents/voyage-ai-input/braga.jpg')
    ],
    [
        "Farm House Lembang adalah destinasi bertema Eropa pedesaan yang menawarkan suasana pedesaan dengan bangunan bergaya Belanda dan interaksi langsung dengan hewan ternak.",
        PIL.Image.open('documents/voyage-ai-input/farm-house.jpg')
    ],
    [
        "Gedung Sate adalah ikon arsitektur kolonial Belanda yang terkenal dengan menara tengahnya yang menyerupai bentuk sate, kini menjadi pusat pemerintahan Jawa Barat.",
        PIL.Image.open('documents/voyage-ai-input/gedung-sate.jpg')
    ],
    [
        "Kawah Putih Ciwidey adalah danau kawah vulkanik dengan air berwarna kehijauan yang dramatis, dikelilingi oleh pemandangan alam pegunungan yang sejuk dan memukau.",
        PIL.Image.open('documents/voyage-ai-input/kawah-putih.jpg')
    ],
    [
        "Gunung Tangkuban Perahu adalah gunung berapi aktif yang menyuguhkan pemandangan kawah utama yang luas dan asap belerang yang terus mengepul, menjadi atraksi alam utama di utara Bandung.",
        PIL.Image.open('documents/voyage-ai-input/tangkuban-parahu.jpg')
    ]
]
```

```python
result = vo.multimodal_embed(inputs, model="voyage-multimodal-3")
```
