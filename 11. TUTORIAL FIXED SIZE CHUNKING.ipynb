```python
from chonkie import TokenChunker
from pathlib import Path
```

```python
text = Path("documents/bandung.md").read_text(encoding="utf-8")
```

```python
chunker = TokenChunker(
    tokenizer='character', #character, word, gpt2
    chunk_size=500,
    chunk_overlap=50 # 0-1 -> persentase overlap
)
```

```python
chunks = chunker.chunk(text)
```
